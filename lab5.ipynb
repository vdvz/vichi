```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

Константы

```python
x0 = 0;
x1 = 10;
n = 20;
n2 = 40;
h = (x1 - x0) / n;
h2 = (x1 - x0) / n2;

n = n + 1
n2 = n2 + 1
```

```python
def drw(y1, y2, _xmax=10, _ymax=10):
    d = len(y1)
    x = [i for i in range(0, d)]
    fig, ax = plt.subplots(figsize=(10, 6))
    #ax.plot(g1, y, marker='o', c='r')
    ax.plot(x, y1, 'ro')
    #ax.plot(g1_2, y2, marker='o', c='b')
    ax.plot(x, y2, 'bx')
    #ax.scatter(x=g1, y=y, marker='o', c='r', edgecolor='b')
    ax.grid(True)
    ax.set_ylabel('Y')
    ax.set_xlabel('X')
    ax.set_ylim(ymin=0, ymax = _ymax)
    ax.set_xlim(xmin=0, xmax = _xmax)
    fig.tight_layout()
    plt.show()
```

```python
def drw3(y1, y2, y3, _xmax=10, _ymax=10):
    if len(y1)!=len(y2)!=len(y2):
        raise
    d = len(y1)
    x = [i for i in range(0, d)]
    fig, ax = plt.subplots(figsize=(10, 6))
    #ax.plot(g1, y, marker='o', c='r')
    ax.plot(x, y1, 'ro')
    ax.plot(x, y2, 'bo')
    #ax.plot(g1_2, y2, marker='o', c='b')
    ax.plot(x, y2, 'bx')
    #ax.scatter(x=g1, y=y, marker='o', c='r', edgecolor='b')
    ax.grid(True)
    ax.set_ylabel('Y')
    ax.set_xlabel('X')
    ax.set_ylim(ymin=0, ymax = _ymax)
    ax.set_xlim(xmin=0, xmax = _xmax)
    fig.tight_layout()
    plt.show()
```

```python
g1 = []
g2 = []
for i in range(0,n):
    g1.insert(i, math.exp(i*h))
    g2.insert(i, math.sin(i*h))
```

```python
g1_2 = []
g2_2 = []
for i in range(0, n2):
    g1_2.insert(i, math.exp(i*h2))
    g2_2.insert(i, math.sin(i*h2))
```

Экспонента 1-го порядка аппроксимации на сетке n=20

```python
y_k1 = []
y_k1.insert(0,1)
for i in range(0, n-1):
    y_k1.insert(i+1, y_k1[i] + h*g1[i])
    
err = 0;
for k in range(0,n):
    err += abs(math.exp(k * h) - y_k1[k]) * h;

print(err)

drw(y_k1, g1, 10, 100)
```

```python
y2_k1 = []
y2_k1.insert(0, 1)
for i in range(0, n2-1):
    y2_k1.insert(i+1, y2_k1[i] + h2*g1_2[i])
    
    
err = 0;
for k in range(0, n):
    err += abs(math.exp(2 * k * h2) - y2_k1[2 * k]) * h2

print(err)

drw(y2_k1, g1_2)
```
