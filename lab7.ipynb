```python
import math
import time
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

n = 100
a = 10
max_time = 5.0
max_x = 5.0
tau = max_time/n
h = max_x/n
x_ = []
for_draw = []
y_ = []
r = a*tau/h


def func1(x):
    if x<=4 and x>=1:
        return math.sin(math.pi/3*(x-1))
    return 0

def func2(x):
    if x<0 and x>=-1:
        return 2
    if x<=1 and x>=0:
        return 1

for i in range(0, n):
    x_.append(i*h)
    y_.append(func1(x_[i]))
    
print(x_)
print(y_)
```

```python
def drw(y1, y2, _xmax=10, _ymax=10):
    d = len(y1)
    x = [i for i in range(0, d)]
    fig, ax = plt.subplots(figsize=(10, 6))
    #ax.plot(g1, y, marker='o', c='r')
    ax.plot(x, y1, 'ro')
    #ax.plot(g1_2, y2, marker='o', c='b')
    ax.plot(x, y2, 'bx')
    #ax.scatter(x=g1, y=y, marker='o', c='r', edgecolor='b')
    ax.grid(True)
    ax.set_ylabel('Y')
    ax.set_xlabel('X')
    ax.set_ylim(ymin=0, ymax = _ymax)
    ax.set_xlim(xmin=0, xmax = _xmax)
    fig.tight_layout()
    plt.show()
    
def drw1(y1, _xmax=5, _ymax=10):
    d = len(y1)
    x = [i*h for i in range(0, d)]
    fig, ax = plt.subplots(figsize=(10, 6))
    #ax.plot(g1, y, marker='o', c='r')
    ax.plot(x, y1, 'ro')
    #ax.plot(g1_2, y2, marker='o', c='b')
    ax.grid(True)
    ax.set_ylabel('Y')
    ax.set_xlabel('X')
    ax.set_ylim(ymin=0, ymax = _ymax)
    ax.set_xlim(xmin=0, xmax = _xmax)
    fig.tight_layout()
    plt.show()
```

```python
for_draw = []
for_draw.append(y_)
for iter_ in range(0, n):    
    y = for_draw[iter_]
    #print(y)
    new_y = []
    y.insert(0, 0)
    for j in range(1, n):
        new_y.insert(1, y[j]*(1-r) + y[j-1]*r)
    for_draw.insert(iter_+1, new_y)
        
#print(for_draw)
        
fig = plt.figure()
ax = plt.axes(xlim=(0, 10), ylim=(0, 10))
line, = ax.plot([], [], lw=3)
def init():
    line.set_data([], [])
    return line,

def animate(i):
    i = i%(n)
    x = [i*h for i in range(0, len(for_draw[i]))]
    y = for_draw[i]
    #print(y)
    line.set_data(x, y)
    return line,
 
anim = FuncAnimation(fig, animate, frames=200, interval=10, blit=True)



for i in range(0, n):
    drw1(for_draw[i])
```
